IMPORTING LIBRARIES

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#Store the website in variable
website = 'https://www.flipkart.com/search?q=laptops&otracker=search&otracker4=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&page=1'
```

```python
#Get Request
response = requests.get(website)
```

```python
#Status Code
response.status_code
```

```python
#SoupObject
soup = BeautifulSoup(response.content,'html.parser')
```

```python
soup
```

```python
#RESULTS
results = soup.find_all('div',{'class':'_2kHMtA'})
```

```python
len(results)
```

```python
#WHAT WE ARE TRYING TO FIND
#1-NAME
#2-PRICE
#3-RATING
```

```python
#NAME
results[0].find('div',{'class':'_4rR01T'}).get_text()
```

```python
#PRICE
results[0].find('div',{'class':'_30jeq3 _1_WHN1'}).get_text()
```

```python
#RATING
results[0].find('div',{'class':'_3LWZlK'})
```

```python
#PUT EVERYTHING INSIDE A FOR LOOP
name = []
price = []
rating = []


for result in results:
    
    
    #name
    try:
        name.append(result.find('div',{'class':'_4rR01T'}).get_text().strip())
    except:
        name.append('n/a')
    
    
    #price
    try:
        price.append(result.find('div',{'class':'_30jeq3 _1_WHN1'}).get_text().strip())
    except:
        price.append('n/a')
        
        
    #rating
    try:
        rating.append(result.find('div',{'class':'_3LWZlK'}).get_text().strip())
    except:
        rating.append('n/a')
```

```python
#Create Pandas DataFrame
lap_tops = pd.DataFrame ({'Name':name,'Price':price,'Rating':rating})
```

```python
lap_tops
```

```python
#DATA CLEANING
lap_tops['Price'] = lap_tops['Price'].apply(lambda x:x.strip('₹'))
```

```python
lap_tops
```

```python
lap_tops.to_csv('laptops2.csv', index = False, encoding = 'utf-8')
```

PAGINATION

```python
name = []
price = []
rating = []



for i in range(1,52):
    
    # website in a variable
    website = 'https://www.flipkart.com/search?q=laptops&otracker=search&otracker4=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&page=' +str(i)
    
    
    #Get Request
    response = requests.get(website)
    
    
    #SoupObject
    soup = BeautifulSoup(response.content,'html.parser')
    
    
    #RESULTS
    results = soup.find_all('div',{'class':'_2kHMtA'})
    
    
    for result in results:
        
        
        #name
        try:
            name.append(result.find('div',{'class':'_4rR01T'}).get_text().strip())
        except:
            name.append('n/a')
        
        
        #price
        try:
            price.append(result.find('div',{'class':'_30jeq3 _1_WHN1'}).get_text().strip())
        except:
            price.append('n/a')
        
        
        
        #rating
        try:
            rating.append(result.find('div',{'class':'_3LWZlK'}).get_text().strip())
        except:
            rating.append('n/a')
    
    
    
```

```python
#Create Pandas DataFrame
lap_tops = pd.DataFrame ({'Name':name,'Price':price,'Rating':rating})
```

```python
lap_tops
```

```python
lap_tops.to_csv('CompleteLaptops.csv',index = False,encoding = 'utf-8')
```
